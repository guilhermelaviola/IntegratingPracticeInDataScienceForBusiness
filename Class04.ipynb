```python
# Importing all the necessary libraries:
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Configuring the style:
sns.set(style='whitegrid')
```

```python
# Creating an example DataFrame:
df = pd.DataFrame({
    'Sales': [153, 210, 263, 316, 374, 422],
    'Expenses': [107, 149, 209, 257, 308, 368]
})
```

```python
# Creating an histogram:
plt.figure(figsize=(8,6))
sns.histplot(df['Sales'], kde=True, color='skyblue')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Creating a box plot:
plt.figure(figsize=(8, 6))
sns.boxplot(data=df['Expenses'], color='lightgreen')
plt.title('Expenses Distribution')
plt.xlabel('Expenses')
plt.show()
```

```python
# Creating a bar plot:
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [153, 210, 263, 316, 374, 422]

plt.figure(figsize=(10, 6))
plt.bar(months, sales, color='skyblue')
plt.title('Mpnthly Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
# Creating a pie plot:
segments = ['Electronics', 'Clothes', 'Food', 'Hardware']
participation = [30, 25, 35, 10]

plt.figure(figsize=(8, 6))
plt.pie(participation, labels=segments, autopct='%1.1f%%', colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'])
plt.title('Market Participation by Segment')
plt.show()
```

```python
# Creating a line plot:
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [175, 205, 238, 364, 452, 342]

df = pd.DataFrame

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Months', y='Sales', marker='o', color='skyblue')
plt.title('Mpnthly Sales Trend')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```
